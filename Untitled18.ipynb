```python
final_dataframe = df[[ 'Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataframe['Current_year']=2020
final_dataframe['Final_Year']=final_dataframe['Current_year']-final_dataframe['Year']
final_dataframe.head()
```

```python
final_dataframe.drop(['Year'],axis=1,inplace=True)
final_dataframe.drop(['Current_year'],axis=1,inplace=True)
final_dataframe.head()
```

```python
final_dataframe = pd.get_dummies(final_dataframe,drop_first=True)
```

```python
# Dependent Feature
x = final_dataframe.iloc[:,1:]
```

```python
### Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x,y)
```

```python
#Plotting graph of featue importance for better visualization
feat_imp = pd.Series(model.feature_importances_, index=x.columns)
feat_imp.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
import numpy as np

# Number of trees in the random forest
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]

# Number of features to consider at every split
max_features = ['auto','sqrt']

#Maximum number of leavels in tress
max_depth = [int(x) for x in np.linspace(5,30,num=6)]
#max_depth.append(None)

# Minimum number of samples requires to splt a node
min_samples_split = [2,5,10,15,100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create random grid
random_grid = {'n_estimators': n_estimators,
                'max_features':max_features,
                'max_depth':max_depth,
                'min_samples_split':min_samples_split,
                'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
# Use the random grid to search for the best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring = 'neg_mean_squared_error', n_iter=10, cv=5, verbose=2, random_state=42, n_jobs=1)
```

```python
rf_random.fit(x_train, y_train)
```

```python
y_pred = rf_random.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics 
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
